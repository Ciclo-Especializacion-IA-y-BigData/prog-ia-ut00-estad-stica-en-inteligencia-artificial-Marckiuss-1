## 🔹 ¿Por qué es fundamental la estadística en Machine Learning?

La estadística y la probabilidad son pilares **críticos** de la Inteligencia Artificial (IA). En **IA** y **Machine Learning (ML)** trabajamos con **vastas cantidades de datos** (a menudo denominados *big data*):
*   Los **likes** en redes sociales.
*   Los **precios** de las casas en una ciudad.
*   Las **notas** de los alumnos de una clase.
*   Las **palabras** más usadas en un texto.

Antes de usar estos datos en un modelo inteligente, necesitamos **entenderlos y describirlos bien**. La estadística actúa como la **primera herramienta matemática** para el procesamiento de datos:

*   **Resumir y Describir:** Necesitamos conocer las propiedades estadísticas del conjunto de datos: ¿cuáles son los **valores más comunes** (media)? ¿Cómo se dispersan los datos? Medidas como la **varianza y la desviación típica** son esenciales para comprender la distribución de los datos y cuantificar la incertidumbre.
*   **Identificar Anomalías:** ¿Hay datos extraños o muy diferentes al resto (*outliers*)? La estadística nos permite identificar estos puntos atípicos, típicamente aquellos que se desvían más de dos **desviaciones típicas** de la media.
*   **Relaciones y Generalización:** ¿Dos variables están relacionadas (ejemplo: más horas de estudio → mejores notas)? El *Machine Learning* busca modelos que puedan **generalizar** a partir de estos patrones aprendidos para predecir resultados en datos nuevos, un proceso fundamentalmente respaldado por la **inferencia estadística**.
*   **Preparación de Datos:** Se utilizan técnicas estadísticas de **normalización** o **estandarización** para ajustar los datos a una media de cero y una desviación típica de uno. Esto asegura que todas las características tengan una escala similar y que ninguna domine el proceso de aprendizaje debido a su magnitud.

La estadística es clave porque nos ayuda a **resumir, describir y explorar** la información antes de entrenar un modelo, transformando los datos brutos en conocimiento útil.

---

## 🔹 Relación con los modelos de ML

La probabilidad y la estadística son críticas en las distintas fases de los modelos de IA:

*   **En modelado y predicción:** Los algoritmos de IA a menudo se basan en **modelos probabilísticos**. En esencia, al entrenar un modelo, se está **estimando la distribución de probabilidad** subyacente de los datos.
    *   En **regresión**, usamos estadística (como la regresión logística o Gaussiana) para medir cómo una variable depende de otra.
    *   En **clasificación**, necesitamos entender la **distribución de clases** para estimar la probabilidad de que una entrada pertenezca a una categoría específica. Por ejemplo, la **Distribución de Bernoulli** es clave para clasificar datos binarios (sí/no o spam/no spam).
*   **En la validación de modelos:** Aplicamos conceptos estadísticos para evaluar la **fiabilidad** de los resultados.
    *   El **Error Cuadrático Medio (MSE)** es una función de pérdida común que es una medida estadística que cuantifica la **varianza de los errores** de predicción. Un MSE menor indica que las predicciones del modelo son mejores y tienen menor variabilidad.
    *   Se aplican otras medidas estadísticas como *accuracy* (precisión), *precision* y *recall* para evaluar el rendimiento.

📌 **Sin estadística, un modelo puede parecer “mágico”, pero no sabremos si realmente funciona bien, si sus resultados son fiables o si está simplemente memorizando los datos de entrenamiento (*overfitting*)**.

# 🎲 Capítulo 2: Probabilidad básica: El Lenguaje de la Incertidumbre

La **probabilidad** es el **marco matemático que trata de la verosimilitud (likelihood) de que ocurran los eventos** en un experimento o sistema aleatorio. Nos proporciona las herramientas para **cuantificar la incertidumbre**.

Se expresa como un número real que siempre se encuentra entre **0 y 1**:

*   **0** significa que el evento es **imposible** y nunca sucederá (ej. sacar un 8 en un dado de 6 caras).
*   **1** significa que el evento es **cierto** o seguro, y siempre ocurrirá.
*   Valores intermedios ($0 < P(E) < 1$) indican que el evento es **incierto**. Por ejemplo: $0.5$ (o 50%) indica igual probabilidad de ocurrencia que de no ocurrencia.

---

## Ejemplos de Eventos y Probabilidad en ML

La probabilidad es fundamental en la Inteligencia Artificial (IA) ya que los sistemas a menudo tienen que operar bajo **incertidumbre o información incompleta**:

1.  **Moneda (Clasificación Binaria):**
    *   Cara: 50% (0.5)
    *   Cruz: 50% (0.5)
    *   *Concepto ML:* Un problema de **clasificación binaria** (como spam o no spam) puede modelarse con la **Distribución de Bernoulli**, que describe un suceso con solo dos posibles resultados.

2.  **Dado de 6 caras (Variables Discretas):**
    *   Probabilidad de sacar un 3: 1 de 6 $\rightarrow \frac{1}{6} \approx 0.166$.
    *   Probabilidad de sacar un número par: 3 de 6 $\rightarrow \frac{3}{6} = 0.5$.
    *   *Concepto ML:* El resultado de un lanzamiento de dado es un ejemplo de **variable aleatoria discreta**, que toma un número finito o contable de valores.

3.  **Clima (Frecuencia relativa):**
    Si en León suele llover 100 días al año, la probabilidad de que un día cualquiera llueva es:
    $$\frac{100}{365} \approx 0.27 \rightarrow 27\%$$

---

## Cómo calcular probabilidades simples (Interpretación Frecuentista)

El cálculo de probabilidades simples que asume que todos los resultados son **igualmente probables** se conoce como la **Definición Clásica o de Laplace**:

$$
P = \frac{\text{Casos favorables}}{\text{Casos posibles}}
$$

Ejemplo:
En una baraja española (40 cartas), la probabilidad de sacar un **as** es:
$$P(\text{As}) = \frac{4}{40} = 0.1 \rightarrow 10\%$$

### 📌 El Enfoque Frecuentista y la Ley de los Grandes Números

Esta definición clásica se relaciona con la **interpretación frecuentista de la probabilidad**: la probabilidad de un evento se describe como su **frecuencia relativa** entre un número infinito de intentos.

El **valor límite** de esta frecuencia relativa, a medida que el número de intentos tiende a infinito, es la verdadera probabilidad subyacente. Este principio es formalizado por la **Ley de los Grandes Números**.

$$
P(\text{evento}) = \lim_{n \to \infty} \frac{\text{número de ocurrencias}}{\text{número de intentos}}
$$

### Ejemplo en Python: Probabilidad Teórica vs. Experimental

Vamos a mostrar cómo la **probabilidad teórica** (0.5) se acerca a la **probabilidad experimental** (lo que obtenemos al simular), ilustrando el concepto de la **Ley de los Grandes Números**.





`````python
import random

# Lanzamos una moneda 10 veces
resultados = [random.choice(["cara", "cruz"]) for _ in range(10)]
print("Resultados:", resultados)

# Probabilidad experimental de cara
prob_cara = resultados.count("cara") / len(resultados)
print("Probabilidad aproximada de cara:", prob_cara)
`````

`````python
import numpy as np
from scipy.stats import bernoulli
import matplotlib.pyplot as plt

# Simulamos 1000 lanzamientos de una moneda (Bernoulli con p=0.5)
n_tosses = 1000
# 0 para cruz, 1 para cara
coin_tosses = [bernoulli.rvs(p=0.5) for _ in range(n_tosses)]

# Calculamos la frecuencia relativa (promedio) acumulada en cada paso
averages = [np.mean(coin_tosses[:k+1]) for k in range(n_tosses)]

# Representación
with plt.style.context("seaborn-v0_8"):
    plt.figure(figsize=(10, 5))
    plt.title("Frecuencia relativa de lanzamientos de moneda")
    plt.xlabel("Número de lanzamientos")
    plt.ylabel("Frecuencia relativa")

    # Plotea los promedios (frecuencia experimental)
    plt.plot(range(n_tosses), averages, linewidth=3)

    # Plotea la probabilidad teórica (0.5)
    plt.plot([-100, n_tosses+100], [0.5, 0.5], c="k")
    plt.xlim(-10, n_tosses+10)
    plt.ylim(0, 1)
    plt.show()
`````

# 📊 3 Estadística descriptiva

La **estadística descriptiva** constituye la primera herramienta matemática utilizada por los sistemas de IA. Su función es **resumir y entender los datos**, permitiendo a los modelos transformar datos brutos en percepciones significativas.

## 1. Medidas de tendencia central: Media, Mediana y Moda

Cuando tenemos un **grupo de números**, hay tres valores que suelen interesarnos mucho:  
- **Media (Mean)** → el promedio.  
- **Mediana (Median)** → el valor del medio.  
- **Moda (Mode)** → el valor que más se repite.  

Estos valores nos ayudan a **resumir la información** de un conjunto de datos sin tener que mirar cada número por separado.

---

## Ejemplo: velocidades de coches
Hemos registrado la velocidad (km/h) de 13 coches:

```
[99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]

```

---

### 🔹 La Media (Mean) o Valor Esperado ($\bar{x}$ o $\mu$)

La media es el promedio aritmético y es la medida de tendencia central más fundamental. En el contexto de la probabilidad y el Aprendizaje Automático, la media corresponde al **valor esperado** ($\mathbb{E}[X]$), que representa el valor promedio de la variable aleatoria a largo plazo.

La fórmula de la media para una población finita de $n$ datos, $x_1, x_2, \ldots, x_n$, se define como:

$$\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i$$

En nuestro caso:

$
\text{Media} = \frac{99+86+87+88+111+86+103+87+94+78+77+85+86}{13} = 89.77
$


**Relevancia en IA (Cálculo Multivariable):**

*   La media es un valor central utilizado para resumir datos y establecer una "línea base".
*   Las desviaciones respecto a la media se utilizan para calcular la **Varianza** y la **Desviación Estándar**, que son medidas cruciales para cuantificar la incertidumbre en las predicciones de los modelos. Por ejemplo, el **Error Cuadrático Medio (MSE)**, una función de pérdida esencial en la regresión lineal, mide la varianza de los errores de predicción.


En Python podemos calcularla con `numpy.mean()`.



---

### 🔹 Mediana (Median)
La Mediana es el valor central de un conjunto de datos ordenados [Query]. A diferencia de la Media, la Mediana no se ve afectada por los **valores atípicos (*outliers*)**, que son valores significativamente distintos al resto.
```

77, 78, 85, 86, 86, 86, 87, 87, 88, 94, 99, 103, 111

```

Aquí, la mediana es **87**, porque está en el centro.  
👉 Importante: ¡los datos **deben estar ordenados** primero!

Si hay un número par de datos, la mediana será la **media de los dos del centro**.  
Ejemplo con 12 datos:

```

77, 78, 85, 86, 86, 86, 87, 87, 94, 98, 99, 103

```

Los dos del medio son 86 y 87 → $(86+87)/2 = 86.5$.

---

### 🔹 Moda (Mode)
La Moda es el valor que aparece con mayor frecuencia en el conjunto de datos

En nuestro caso:

```

99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86

```

El valor **86** aparece **3 veces**, así que la moda es **86**.

---

Ejemplo en Python con sueldos


`````python
import numpy as np
from scipy import stats

# Datos: la mayoría sueldos normales y uno muy alto
sueldos = [1200, 1300, 1250, 1400, 1350, 1280, 1500, 1250, 1300, 30000]

# Medidas
media = np.mean(sueldos)
mediana = np.median(sueldos)
moda = stats.mode(sueldos, keepdims=True)[0][0]

print("Media:", media)
print("Mediana:", mediana)
print("Moda:", moda)

`````

📌 **Conclusión**

* La **media** nos da una idea del valor promedio.
* La **mediana** nos dice cuál es el valor central (muy útil si hay valores extremos que distorsionan la media).
* La **moda** muestra el valor más frecuente.

El uso de estas medidas es indispensable para **entender los datos antes de entrenar un modelo**, lo que se traduce en técnicas de preprocesamiento esenciales para la fase inicial del ML:

*   **Limpieza de Datos:** Antes de que un algoritmo pueda aprender, debe comprender las propiedades estadísticas de los datos, incluyendo su distribución y varianza. Un buen manejo de la estadística permite tareas de preprocesamiento como la limpieza de datos y la remoción de valores atípicos.
*   **Normalización y Estandarización:** Estas medidas se utilizan para aplicar **técnicas de normalización** (como la estandarización por puntuación z), que son herramientas matemáticas que ajustan los datos para que tengan una media de cero y una desviación estándar de uno. Este proceso asegura que todas las características estén en una escala similar, impidiendo que una característica, debido a su magnitud, **domine el proceso de aprendizaje**.
*   **Funciones de Pérdida (Loss Functions):** Estos conceptos descriptivos se integran directamente en las funciones de optimización. Por ejemplo, el **Error Cuadrático Medio (MSE)**, una función de pérdida común en regresión, se define en términos del **valor esperado o media** y cuantifica la varianza de los errores de predicción.

En esencia, la estadística descriptiva sienta las bases matemáticas para que los algoritmos de **Cálculo Multivariable** (como el descenso de gradiente) y **Álgebra Lineal** puedan operar de manera eficiente y precisa en la búsqueda de patrones y la optimización de las redes neuronales.

## 🔹 Rango

El rango, también llamado span, es la distancia entre el mínimo y el máximo de una distribución, es decir, la distancia entre el valor más pequeño y el más grande.


![](https://numiqo.es/assets/tutorial/Rango.png)


### Ejemplo

`````python
import numpy as np

notas = [5, 7, 8, 6, 7, 10, 4, 7, 9, 6]
rango = np.max(notas) - np.min(notas)

print("Rango:", rango)
`````

📌 El rango muestra la amplitud total de los datos.


## 🔹 Desviación Típica ($\sigma$)

La **desviación típica (o estándar)** es la **raíz cuadrada de la varianza** . Esta medida es preferida para la interpretación práctica porque está expresada en las **mismas unidades** que los datos originales .

*   Una **desviación baja** indica que los datos están agrupados muy **cerca de la media** .
*   Una **desviación alta** indica que los datos están **muy dispersos** o separados entre sí.

![](https://datatab.es/assets/tutorial/Desviaci%C3%B3n_est%C3%A1ndar.png)

---

## 🔹 Varianza ($\text{Var}[X]$ o $\sigma^2$)

La **varianza** mide la desviación cuadrática promedio de los datos respecto a la media. En el contexto probabilístico, la varianza es el **valor esperado de la desviación al cuadrado** de una variable aleatoria ($X$) respecto a su media ($\mu$):

$$\text{Var}[X] = \mathbb{E}[(X - \mu)^2]$$

La fórmula de la varianza para un conjunto de datos finito ($n$ observaciones) es:

$$\sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2$$

El propósito de elevar al cuadrado las diferencias es eliminar los signos negativos y penalizar las desviaciones grandes con mayor peso, una técnica que se traslada directamente a las funciones de pérdida en ML (como se verá a continuación).

Como los valores se elevan al cuadrado, el resultado tiene una unidad diferente (la unidad al cuadrado) que los valores originales. Por lo tanto, es difícil relacionar los resultados.

---

## Varianza frente a desviación estándar

Así pues, la diferencia entre la varianza y la desviación estándar del parámetro de dispersión es que la desviación estándar mide la distancia media respecto a la media y la varianza mide la distancia media al cuadrado respecto a la media. En otras palabras, la varianza es la desviación estándar elevada al cuadrado y la desviación estándar es la raíz de la varianza.

Sin embargo, esta elevación al cuadrado da lugar a un ratio difícil de interpretar, ya que la unidad no se corresponde con los datos originales. Por este motivo, es aconsejable utilizar siempre la desviación típica para describir una muestra, ya que facilita la interpretación.

---


`````python
puntuaciones = [4, 5, 5, 8, 9, 12, 14, 16, 17, 20]

varianza = np.var(puntuaciones)
desviacion_tipica = np.std(puntuaciones)

print("Varianza:", varianza)
print("Desviación típica:", desviacion_tipica)
`````


## 3.2. El Impacto de la Dispersión en Machine Learning (ML)

En **Machine Learning**, la desviación típica y la varianza son indispensables para la fase de preprocesamiento de datos y para la etapa de optimización del modelo [Query]. La estadística es una de las primeras herramientas matemáticas utilizadas en el procesamiento de datos de IA.

### A. Función de Pérdida y Optimización

Las medidas de dispersión se encuentran en el corazón de la optimización de los modelos de IA, que es el proceso de ajustar los parámetros (pesos) para minimizar una función de error.

*   **Error Cuadrático Medio (MSE):** La varianza está directamente relacionada con una de las funciones de pérdida más comunes en los modelos de regresión, el **Error Cuadrático Medio (MSE)**. El MSE esencialmente mide la **varianza de los errores** de predicción. Un MSE bajo (es decir, una varianza baja del error) indica que las predicciones del modelo son mejores.
*   **Cuantificación de la Incertidumbre:** La varianza y la desviación típica son vitales para **evaluar el rendimiento del modelo** y **cuantificar la incertidumbre** en las predicciones de la IA. Una baja varianza en las predicciones sugiere una alta confianza del modelo, mientras que una alta varianza indica incertidumbre.

### B. Normalización y Estandarización de Datos

La dispersión es clave en el proceso de **preprocesamiento** de datos, una etapa crucial **antes de entrenar un modelo**:

*   **Estandarización (Normalización Z-score):** Para que un modelo de ML (especialmente las redes neuronales) aprenda de manera efectiva, es común estandarizar los datos para que tengan una **media de cero y una desviación estándar de uno** ($\sigma=1$).
*   **Prevención de Dominio de Características:** Este proceso matemático (que es una transformación lineal) asegura que todas las características estén en una escala similar, **impidiendo que una sola característica domine el proceso de aprendizaje** debido a su magnitud (por ejemplo, si el "salario" tiene valores de miles y la "edad" valores de decenas).

![](https://www.mygreatlearning.com/blog/wp-content/uploads/2025/03/data-after-standardization.jpg.webp)


### C. Detección de Valores Atípicos (*Outliers*)

La Desviación Típica sirve como una métrica robusta para identificar si una variable tiene valores "normales" o si presenta **outliers**.

*   Los **valores atípicos** son puntos de datos que se desvían significativamente del resto.
*   Típicamente, cualquier punto que se encuentre a más de **2 desviaciones estándar** ($\pm 2\sigma$) de la media se considera un valor atípico.

![](https://statistics.laerd.com/spss-tutorials/img/lr/outliers.png)

### D. Reducción de Dimensionalidad (PCA)

La varianza es el concepto central de técnicas avanzadas como el **Análisis de Componentes Principales (PCA)**.

*   PCA, que es crucial para manejar datos de alta dimensión (como imágenes o texto), primero requiere que los datos sean estandarizados (usando media y desviación típica).
*   PCA busca las direcciones (autovectores) en el espacio de datos donde la **varianza es máxima**. Los autovalores correspondientes indican la magnitud de esa varianza. Al seleccionar solo los autovectores con las mayores varianzas (los autovalores más grandes), PCA reduce la dimensionalidad mientras conserva la información más significativa.

## 🔹 Cuartiles

Los cuartiles dividen los datos en cuatro partes, lo más iguales posible. Para el cálculo de los cuartiles, los datos deben estar ordenados del valor más pequeño al más grande.

- Cuartil (Q1): El valor medio entre el valor más pequeño (mínimo) y la mediana.
- Cuartil (Q2): La mediana de los datos, es decir, el 50% de los valores son menores y el 50% de los valores son mayores.
- Cuartil (Q3): El valor medio entre la mediana y el valor más grande (máximo).

1. Calcula **Q1, Q2 y Q3** con `np.percentile`.
2. Halla el **IQR = Q3 − Q1**.
3. Calcula los límites para detectar outliers:

   * Límite inferior = Q1 − 1.5·IQR
   * Límite superior = Q3 + 1.5·IQR
4. Determina qué valores de CI son atípicos.

`````python
q1 = np.percentile(ci, 25)
q2 = np.percentile(ci, 50)
q3 = np.percentile(ci, 75)
iqr = q3 - q1

l_inf = q1 - 1.5 * iqr
l_sup = q3 + 1.5 * iqr

print(f"Q1: {q1}, Q2: {q2}, Q3: {q3}, IQR: {iqr}")
print(f"Límites: [{l_inf}, {l_sup}]")

outliers = [x for x in ci if x < l_inf or x > l_sup]
print(f"Valores atípicos detectados: {outliers}")
`````

## 🔹 Histogramas

Un **histograma** es un gráfico que nos permite ver cómo se distribuyen los datos en intervalos (llamados **bins**).  

### 1. Vínculo con la Probabilidad y el Cálculo

El concepto fundamental que subyace al histograma es la **función de densidad de probabilidad (PDF)**, un elemento central de la probabilidad y la estadística.

*   **Bins y PDF:** Los histogramas dividen el rango de valores de una variable en **intervalos (bins)** . La altura de cada barra de frecuencia es una aproximación visual de la función de densidad de probabilidad. Para las variables continuas, la probabilidad se encuentra integrando la PDF sobre un rango de valores. Un histograma bien construido intenta emular la forma de esta función matemática subyacente.
*   **Distribuciones Notables:** Los histogramas permiten visualizar si la distribución de una característica se ajusta a distribuciones teóricas comunes, como la **Distribución Normal (Gaussiana)** o "Campana de Gauss", crucial para muchos modelos de ML que asumen un comportamiento normal.

### 2. Detección de Patrones y Anomalías (Outliers)

La detección de **valores atípicos (*outliers*)** es crítica en el preprocesamiento de datos, ya que estos pueden distorsionar significativamente la **Media** y la **Varianza**, afectando la precisión del modelo.

*   **Concentraciones:** Las concentraciones más altas en el histograma reflejan las **medidas de tendencia central** (Moda y Media) [Query].
*   **Detección Visual de Outliers:** Los *outliers* se manifiestan visualmente en el histograma como barras de muy baja frecuencia ubicadas lejos de la concentración principal de los datos. Esta detección visual apoya el análisis estadístico más riguroso, que define un *outlier* típicamente como un punto a más de **2 desviaciones estándar ($\sigma$)** de la media.
*   **Asimetría (*Skewness*):** Los histogramas revelan si la distribución es simétrica (como la Gaussiana) o asimétrica. Si una distribución está sesgada, la media es un descriptor menos fiable que la mediana.

En **Machine Learning**, estas medidas se usan mucho para **entender los datos antes de entrenar un modelo**. La visualización mediante histogramas es un paso previo indispensable para la optimización de los modelos:

*   **Estandarización y Normalización:** La mayoría de los algoritmos de optimización (como el **Descenso de Gradiente**) funcionan mejor si los datos están estandarizados (centrados en una media de cero y con una desviación estándar de uno). Los histogramas permiten verificar visualmente que las características están en una **escala similar**, evitando que una característica "domine el proceso de aprendizaje debido a su magnitud".
*   **Análisis de Características (Feature Analysis):** En aplicaciones que manejan datos de alta dimensión (como el *deep learning*) es crucial analizar las características individuales antes de aplicar técnicas de reducción de dimensionalidad como **PCA** (Análisis de Componentes Principales), que se basa en la **máxima varianza** en los datos. 
# Boxplot (Diagrama de caja)

El **Diagrama de Caja (Boxplot)** es una herramienta gráfica de la estadística descriptiva que resulta inestimable en la fase de **Análisis Exploratorio de Datos (EDA)**, un paso previo e indispensable **antes de entrenar un modelo de Machine Learning** .

Este gráfico es fundamental porque ofrece, en una sola visualización, una perspectiva completa sobre la **dispersión**, la **simetría** y, lo más crítico para los modelos de IA, la presencia de **valores atípicos (*outliers*)**

A continuación, se detalla la composición y la relevancia matemática y de IA del *boxplot*:

### 1. Componentes del Diagrama de Caja

El diagrama de caja se construye a partir de los cuantiles (percentiles) que dividen el conjunto de datos ordenado en cuatro partes iguales, y proporciona una medida de dispersión que es **robusta** ante valores extremos:

![](https://miro.medium.com/v2/resize:fit:1200/1*0MPDTLn8KoLApoFvI0P2vQ.png)

### 2. Ventaja en el Análisis de Datos (EDA)

La principal ventaja del *boxplot* es que **permite ver dispersión, simetría y valores atípicos en un solo gráfico** .

*   **Dispersión y Variabilidad:** La longitud de la caja (IQR) y la extensión de los bigotes ofrecen una visualización inmediata de cuán agrupados o dispersos están los datos.
*   **Simetría:** Si la mediana está en el centro de la caja y los bigotes son de longitud similar, la distribución tiende a ser simétrica (aproximándose a una distribución normal o Gaussiana).

### 3. El Boxplot y la Detección de *Outliers* en Machine Learning

La capacidad del diagrama de caja para identificar valores atípicos es crucial en ML, ya que estos puntos de datos pueden impactar drásticamente el proceso de entrenamiento.

1.  **Distorsión de la Media y la Varianza:** Los valores atípicos son datos que se desvían significativamente. Si no se manejan, los *outliers* pueden afectar gravemente la **media ($\mu$)** y la **Varianza ($\sigma^2$)**, lo cual es problemático porque la optimización en las redes neuronales a menudo se basa en minimizar errores cuadráticos (como el MSE), que dependen de la varianza.
2.  **Métrica de Detección:** El IQR proporciona un método *robusto* para establecer umbrales de atipicidad. Sin embargo, en la práctica de IA, la detección de *outliers* se relaciona comúnmente con la **Desviación Típica ($\sigma$)**, otra medida de dispersión:

    *   Típicamente, cualquier punto de dato que se encuentre a **más de 2 desviaciones estándar ($\sigma$) de la media ($\mu$) se considera un *outlier***.

3.  **Preprocesamiento y Estandarización:** La estadística descriptiva (media y desviación típica) se utiliza para la **estandarización** o **normalización** de los datos, ajustándolos para que tengan una media de cero y una desviación estándar de uno. Este paso es vital para asegurar que todas las características tengan la misma influencia en el modelo y que ninguna característica "domine el proceso de aprendizaje". La visualización con *boxplots* puede guiar al analista sobre si los datos deben ser estandarizados o si los *outliers* deben ser eliminados antes de aplicar esta transformación.

`````python
import matplotlib.pyplot as plt

notas = [4,5,6,7,8,5,6,7,9,10,6,7,5,4,8,9,10,6,7,8,0,20]

plt.boxplot(notas)
plt.title("Boxplot de notas de alumnos con outliers")
plt.show()
`````

📌 Interpretación:

* La **línea central** indica la mediana.
* La **caja** indica el rango intercuartílico (la mayoría de las notas).
* Los **bigotes** se extienden a los valores mínimo y máximo que no son atípicos.
* Los outliners se marcan como puntos fuera de la caja (0 y 20).

5. Detecta los **outliers** estadísticamente (fuera de `[Q1 - 1.5·IQR, Q3 + 1.5·IQR]`).

   👉 Reflexiona: ¿qué significan estos valores extremos en términos de hábitos de sueño? ¿Podrían ser errores de registro o casos reales interesantes?

`````python
import matplotlib.pyplot as plt
import numpy as np

suenio = [6, 7, 8, 7, 6, 9, 5, 8, 7, 6, 7, 8, 10, 4, 12, 7, 6, 5, 8, 9]

#  Nivel básico 
# 1 y 2. Crear Boxplot con etiquetas
plt.boxplot(suenio, vert=False, patch_artist=True, flierprops=dict(markerfacecolor='r', marker='D'))
plt.title("Distribución de horas de sueño")
plt.xlabel("Horas")
plt.show()

#  Nivel intermedio
# 3. Calcular Cuartiles e IQR
q1 = np.percentile(suenio, 25)
q2 = np.percentile(suenio, 50)  # Mediana
q3 = np.percentile(suenio, 75)
iqr = q3 - q1

print(f"Q1: {q1} | Mediana (Q2): {q2} | Q3: {q3} | IQR: {iqr}")

#  Nivel avanzado Reto
# 5. Detectar outliers estadísticamente
print()
print("Detectar outliers estadísticamente")
limite_inf = q1 - 1.5 * iqr
limite_sup = q3 + 1.5 * iqr

outliers = [x for x in suenio if x < limite_inf or x > limite_sup]
print(f"Límites de outliers: [{limite_inf}, {limite_sup}]")
print(f"Outliers detectados: {outliers}")
`````

Un **scatter plot** es un gráfico de puntos donde cada punto representa un par de valores (x, y).  
Sirve para ver si existe relación entre dos variables.  

---

## 🔹 Ejemplo simple

La edad de un coche (x) y su velocidad (y):

## 🔹 Scatter con datos aleatorios grandes

En **Machine Learning**, los datasets pueden tener **miles o millones de valores**.
Cuando no tenemos datos reales, podemos **generar datos simulados** con NumPy.

👉 Ejemplo: crear dos arrays de 1000 valores cada uno, siguiendo distribuciones normales diferentes.

* `x` con **media = 5.0** y **desviación típica = 1.0**.
* `y` con **media = 10.0** y **desviación típica = 2.0**.

`````python
import numpy as np

# 1000 valores simulados con distribución normal
x = np.random.normal(5.0, 1.0, 1000)   # media 5, σ=1
y = np.random.normal(10.0, 2.0, 1000)  # media 10, σ=2

plt.scatter(x, y, alpha=0.5)
plt.title("Scatter plot con 1000 puntos")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
`````

* Los puntos se concentran alrededor de (5,10).
* La nube es más extendida en el eje Y (desviación 2) que en X (desviación 1).

---

✅ **Idea clave**:

* Un **histograma** describe **una variable**.
* Un **scatter plot** describe la relación entre **dos variables**.
* Son básicos en Machine Learning para descubrir **correlaciones** y patrones.

# 🔔 Distribución Normal (Campana de Gauss)

En la vida real, muchos fenómenos siguen una **distribución normal**:  
- Altura de las personas  
- Resultados de exámenes  
- Errores de medida  

👉 Características:
- Los valores se concentran alrededor de la **media**.  
- La probabilidad disminuye a medida que nos alejamos de la media.  
- La gráfica tiene forma de **campana**.  

---

## 🔹 Ejemplo en Python

`````python
import numpy as np
import matplotlib.pyplot as plt

# 100.000 valores con media=5.0 y desviación típica=1.0
x = np.random.normal(5.0, 1.0, 100000)

plt.hist(x, 100)
plt.title("Distribución normal (media=5, σ=1)")
plt.show()
`````

📌 Interpretación:

* La mayoría de valores están entre **4 y 6**.
* El máximo se encuentra en la **media (5.0)**.
* Pocos valores se alejan mucho → los extremos son raros.

---

✅ Esta distribución se conoce como **campana de Gauss** y es fundamental en probabilidad, estadística y Machine Learning.

# 5 Correlación y Covarianza

Cuando trabajamos con **dos variables**, no solo nos interesa su distribución individual,  
sino también **cómo se relacionan entre sí**.

---

## 🔹 Covarianza

- La **covarianza** mide cómo varían dos variables juntas.  
- Si aumenta una y también lo hace la otra → covarianza positiva.  
- Si una aumenta y la otra disminuye → covarianza negativa.  
- Si no hay patrón → covarianza ≈ 0.  

📌 Problema: depende de las unidades, por lo que no es fácil comparar.

---

## 🔹 Correlación

- La **correlación** es una medida estandarizada de la relación entre dos variables.  
- Valores entre **-1 y 1**:  
  - +1 → correlación positiva perfecta.  
  - -1 → correlación negativa perfecta.  
  - 0 → sin relación.  

📌 La correlación es más útil porque no depende de las unidades de medida.

---

## 🔹 Ejemplo: Tamaño de vivienda vs Precio

Queremos analizar si las viviendas más grandes suelen ser más caras.

`````python
import numpy as np
import pandas as pd

# Datos ficticios: tamaño en m² y precio en miles de €
tamanio = [50, 60, 80, 100, 120, 150, 200]
precio  = [100,120,150,180,210,250,320]

# Covarianza con NumPy
cov = np.cov(tamanio, precio, ddof=0)
print("Matriz de covarianza:\n", cov)

# Correlación con NumPy
corr = np.corrcoef(tamanio, precio)
print("\nMatriz de correlación:\n", corr)

# Con Pandas
df = pd.DataFrame({"Tamaño": tamanio, "Precio": precio})
print("\nResumen con Pandas:")
print(df.corr())
`````

---

## 🔎 Explicación de resultados

* La **covarianza** será positiva → cuando aumenta el tamaño, también aumenta el precio.
* La **correlación** estará cerca de **+1** → relación fuerte y positiva.

📌 En un scatter plot veríamos los puntos alineados en forma ascendente.

`````python

import matplotlib.pyplot as plt

plt.scatter(tamanio, precio)
plt.title("Tamaño de vivienda vs Precio")
plt.xlabel("Tamaño (m²)")
plt.ylabel("Precio (miles de €)")
plt.show()
`````

✅ **Idea clave:**

* La **covarianza** nos dice si dos variables suben o bajan juntas.
* La **correlación** nos dice además **qué tan fuerte es esa relación**.
* En Machine Learning, estas medidas nos ayudan a:

  * Seleccionar variables relevantes (ej. si dos variables están muy correlacionadas, puede bastar con una).
  * Entender qué factores influyen en un resultado.
